# Collatz Mapper

The idea is to make a particular visual representation of the [Collatz Conjecture](https://en.wikipedia.org/wiki/Collatz_conjecture "Collatz Conjecture") where we consider a generalized version of the conjecture: $a n + b$ and divisor $d$.

First, Numpy is loaded and some useful functions are loaded from the file `collatz.py`.

```python
import numpy as np
from collatz import collatz_mapper, collatz_plotter
```

```python
amax = 16
bmax = 16
div = 2
nlim = 20

collatz_map = collatz_mapper(amax, bmax, div, nlim)

collatz_plotter(collatz_map, div)
```

## Load a map

```python
#Load Collatz Map
collatz_map = np.loadtxt('maps/collatz_map_a64_b64_div3.colmap')
```

```python
#Plotting the map
collatz_plotter(collatz_map, 3)
```

```python
#If map being loaded was made using C file, it should be transposed
collatz_plotter(collatz_map.transpose(), 3)
```

## Save a map to a text file

```python
#Save collatz map to file
filename = 'collatz_map_a'+str(amax)+'_b'+str(bmax)+'_div'+str(div)+'.colmap'
with open(filename, 'w') as file:
    for row in range(2*amax):
        for column in range(2*bmax):
            file.write(str(collatz_map[row,column])+'\t')
        file.write('\n')
```
